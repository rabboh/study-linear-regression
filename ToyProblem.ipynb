# Linear Regression Toy Problem

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_train = [3, 5, 3.5, 4.25, 4, 4.59, 3.8, 1, 2, 1.6, 2.5, 2.6, 2.7, 1.4]
y_train = [5, 3, 4.5, 3.1, 4, 3.74, 4.4, 8.9, 6.5, 7.5, 6.5, 4.9, 4.93, 8.7]
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title('Plotting our data')
plt.show()
```

It is a univariate linear regression problem.   
We have a very simple toy dataset for this.  
Our model: f(x) = wx+b   
w and b are the parameters of our model (coefficients or weights)  
These parameters are aligned during the training.

b is the y-intercept, and w is the slope of the line.

```python
# Let's play with w and b
w = -2
b = 11
```

```python
def f(x, w, b):
    return w*x+b;

x1 = []
y1 = []
for x in range(6):
    x1.append(x)
    y1.append(f(x, w, b))
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title('Plotting our data')
plt.plot(x1, y1)
plt.show()
```

!! The plot above was set by hand, try-and-fail. We know better.

We should find w and b parameters that makes the cost function minimal. The gradient descent is an algorithm that can find that w and b by minimizing the cost function. It is widely used, not just for the linear regression.

```python
sum_cost = 0
for i, val in enumerate(x_train):
    sum_cost = sum_cost + (f(x_train[i], w, b) - y_train[i])**2
cost = 1/2*len(x_train) * sum_cost
print(cost)
```
