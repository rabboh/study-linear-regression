# Exploration of the Real Estate Dataset

```python
import csv
import random
import matplotlib.pyplot as plt
```

```python
filename = 'dataset.csv'
fields = []
rows = []
count = 0
```

### Headers, row count, example row

```python
with open(filename, 'r') as csvinput:
    inputreader = csv.reader(csvinput)
    fields = next(inputreader)
print(fields)
```

```python
with open(filename, 'r') as csvinput:
    inputreader = csv.reader(csvinput)
    next(inputreader) # we don't need the header here
    for row in inputreader:
        rows.append(row)
print(len(rows))
```

```python
print(random.choice(rows))
```

## Doing some plotting

```python
v_x = []
v_y = []
for row in rows:
    v_x.append(row[2])
    v_y.append(row[7])
plt.scatter(v_x, v_y)
```
